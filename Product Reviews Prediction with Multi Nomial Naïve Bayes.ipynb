The multinomial Naive Bayes classifier is suitable for classification with discrete features (e.g., word counts for text classification). The multinomial distribution normally requires integer feature counts. However, in practice, fractional counts such as tf-idf may also work.

## **Import Library**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Import Data**

```python
data = pd.read_csv("https://raw.githubusercontent.com/YBIFoundation/ProjectHub-MachineLearning/main/Women%20Clothing%20E-Commerce%20Review.csv")
```

## **Describe Data**

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

## **Data Preprocessing**

```python
data.isna().sum()
```

```python
data[data['Review']==""]=np.NaN
```

```python
data['Review'].fillna("No Review",inplace=True)
```

```python
data.isna().sum()
```

```python
data['Review']
```

```python
data['Rating'].value_counts()
```

## **Define Target Variable (y) and Feature Variables (X)**

```python
data.columns
```

```python
x = data['Review']
x
```

```python
y = data['Rating']
y
```

```python
data['Rating'].value_counts()
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.3,stratify=y,random_state=4356)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape,
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(lowercase=True,analyzer="word",ngram_range=(2,3),stop_words="english",max_features=5000)
```

```python
X_train = cv.fit_transform(X_train)
```

```python
cv.get_feature_names_out()
```

```python
X_train.toarray()
```

```python
X_test = cv.fit_transform(X_test)
```

```python
cv.get_feature_names_out()
```

```python
X_test.toarray()
```

## **Modeling**

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

## **Prediction**

```python
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
model.predict_proba(X_test)
```

## **Model Evaluation**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
